```
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

new_df=df.apply(label.fit_transform)
```

```
new_df
```

```
x=new_df.iloc[:,:-1]
```

```
y=new_df.iloc[:,-1]
```

```
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=40)
```

```
from keras import Sequential
from keras.layers import Dense
```

```
import keras
```

```
model=Sequential()
```

```
model.add(Dense(units=8,input_dim=13))  #we adding 8 neurons to this layer  (First layer)
model.add(Dense(units=12,activation='relu')) #second layer  (hiden layer)       relu or tanh
model.add(Dense(units=1,activation='softmax'))  #output layer
```

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])#after adding the layers we are compiling    optimizer tunes the weights  optimizers=sgd,adam
```

```
#Train model

model.fit(x,y,epochs=150,batch_size=100,verbose=1)   
```
