```
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
 import seaborn as sns
```

```
df=pd.read_csv('/content/diabetes (1).csv')
```

```
df
```

```
df.isnull().sum()
```

```
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```
x
```

```
sns.heatmap(df.isnull())
```

```
from keras import Sequential    #importing ann
```

```
from keras.layers import Dense
```

```
import keras  #importing all the keras api's
```

```
model=Sequential()  #initializing the model
```

```
#adding layer

model.add(Dense(units=8,input_dim=8))  #we adding 8 neurons to this layer  (First layer)
model.add(Dense(units=12,activation='relu')) #second layer  (hiden layer)       relu or tanh
model.add(Dense(units=1,activation='sigmoid')) #output layer  one neuron because one cell for output  sigmoid or softmax
```

```
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])#after adding the layers we are compiling    optimizer tunes the weights  optimizers=sgd,adam
```

```
#Train model

model.fit(x,y,epochs=150,batch_size=10,verbose=1)   #verbose means verbal description
                                                    #epoch is iterations
                                                    #768 rows as a batch_size of 10 ,each batch as 10 samples 
                                                    #in 1st iteration it iterates through all the 77 batches and again 2nd iteration through all the 7 batches 
                                                    #after iterating through all the batches it will come out
```
