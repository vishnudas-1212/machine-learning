```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('/content/diabetes.csv')
```

```
df
```

```
df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=(df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan))
```

```
df
```

```
df.fillna(value=df.mean(),inplace=True)
```

```
df
```

```


X=df.iloc[:,:-1]
```

```
y=df.iloc[:,-1]
```

```
#modeling

from sklearn.tree import DecisionTreeClassifier
```

```
model=DecisionTreeClassifier()
```

```
model.fit(X,y)
```

```
y_pred=model.predict(X)
```

```
y_pred
```

```
new_df=df.copy()
```

```
new_df['predicted']=y_pred
```

```
new_df
```

```
from sklearn import tree
```

```
fig=plt.figure(figsize=(100,100))
_=tree.plot_tree(model,feature_names=df.columns,label='all',filled=True)
```
