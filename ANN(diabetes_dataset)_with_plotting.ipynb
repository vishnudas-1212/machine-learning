```
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
 import seaborn as sns
```

```
df=pd.read_csv('/content/diabetes (1).csv')
```

```
df
```

```
df.isnull().sum()
```

```
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```
x
```

```
sns.heatmap(df.isnull())
```

```
from keras import Sequential    #importing ann
```

```
from keras.layers import Dense
```

```
import keras  #importing all the keras api's
```

```
model=Sequential()  #initializing the model
```

```
#adding layer

model.add(Dense(units=8,input_dim=8))  #we adding 8 neurons to this layer  (First layer)
model.add(Dense(units=12,activation='tanh')) #second layer  (hiden layer)       relu or tanh
model.add(Dense(units=1,activation='sigmoid')) #output layer  one neuron because one cell for output  sigmoid or softmax
```

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])#after adding the layers we are compiling    optimizer tunes the weights  optimizers=sgd,adam
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=40)
```

```
#Train model

history=model.fit(x_train,y_train,epochs=500,validation_data=(x_test,y_test),batch_size=10,verbose=1)   #verbose means verbal description
                                                    #epoch is iterations
                                                    #768 rows as a batch_size of 10 ,each batch as 10 samples 
                                                     #in 1st iteration it iterates through all the 77 batches and again 2nd iteration through all the 7 batches 
                                                    #after iterating through all the batches it will come out
```

```
#Training and testing
```

```
print(history.history.keys())
plt.plot(history.history['accuracy'])

plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
