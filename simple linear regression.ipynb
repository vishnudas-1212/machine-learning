```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data

data=pd.read_csv('sat.csv')
```

```python
data
```

```python
#split x and y

x=data[['SAT']]  #single bracket means only one dimensional
```

```python
x
```

```python
y=data[['GPA']]
```

```python
y
```

```python
#EDA(Exploratory data analysis)

plt.scatter(x,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.title("sat score")
plt.show()
```

```python
data.isna()
```

```python
data.isna().sum()  # check missing values
```

```python
#applying ml algorithm

from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()     #initialize model
```

```python
#training model


model.fit(x,y)
```

```python
y_predict=model.predict(x)     
```

```python
#testing phase

model.predict([[1850]])
```

```python
#caluclating b0 and b1


model.coef_
```

```python
model.intercept_
```

```python
#caluclating y hat =b0+b1+x

yhat= model.intercept_ + model.coef_ * x

```

```python
yhat
```

```python
#EDA(Exploratory data analysis)

plt.scatter(x, y)
plt.plot(x,yhat)

plt.xlabel('SAT')
plt.ylabel('GPA')
plt.title("sat score")
plt.show()
```
